```python

#Classification Model
def precision_calc(tp, fp,):
    precision = tp / (tp + fp)
    return precision

def recall_calc(tp, fn):
    recall = tp / (tp + fn)
    return recall

def f1score_calc(precision, recall):
    if precision + recall == 0:
        return 0
    f1score = 2 * (precision * recall) / (precision + recall)
    return f1score

#Calling Calculation
def f1score_model(tp, fp, fn):
    precision = precision_calc(tp, fp)
    recall = recall_calc(tp, fn)
    f1score = f1score_calc(precision, recall)
    print(f"Precision: {precision} | Recall: {recall} | F1 Score: {f1score}")

#Check rules
def validate_input(tp, fp, fn):
    #Input Prompt
    tp = input("please input number for tp");
    fp = input("please input number for fp");
    fn = input("please input number for fn");
    
    try:
        # Convert input strings to integers and check if any of them are non-zero
        if int(tp) != 0 or int(fp) != 0 or int(fn) != 0:
            return validate_not_zero(tp, fp, fn)
        else:
            return False  # All inputs are zero, return False
        
    except ValueError:
        return False  # If conversion to integer fails, return Fal


def validate_not_zero(tp, fp, fn):
    if int(tp) > 0 | int(tp) > 0 | int(tp) > 0:
        return f1score_model(tp, fp, fn)
    else:
        pass


#Run
validate_input(tp, fp, fn)

```
